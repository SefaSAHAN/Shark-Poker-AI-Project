```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df2 = pd.read_csv('same40_pre_flop.csv')

# Set index to cards
df2.set_index('cards', inplace=True)

# Plot line chart
df2.plot(kind='line', marker='o')
plt.title('Winning Probability for 7S 10H')
plt.xlabel('Simulation')
plt.ylabel('Winning Probability')
plt.ylim(0.492, 0.496)

# Add horizontal line at y=0.4938546659938889
plt.axhline(y=0.4938546659938889, color='green', linestyle='--', label='True Win Rate')

# Show plot
plt.legend()
plt.show()
```

```python

import numpy as np

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('same40_pre_flop.csv')

# Extract the winning probabilities as a numpy array
win_probs = df['win_prob'].values

# Calculate the mean
mean = np.mean(win_probs)

# Calculate the standard deviation
std_dev = np.std(win_probs)

# Calculate the minimum and maximum
minimum = np.min(win_probs)
maximum = np.max(win_probs)

# Calculate the median
median = np.median(win_probs)

# Print the results
print("Mean: {:.4f}".format(mean))
print("Standard deviation: {:.4f}".format(std_dev))
print("Minimum: {:.4f}".format(minimum))
print("Maximum: {:.4f}".format(maximum))
print("Median: {:.4f}".format(median))
```

```python
import pandas as pd
import plotly.express as px

# Load the data from the CSV files
pre_flop_df = pd.read_csv('pre_flop.csv')
normal_df = pd.read_csv('normal1.csv')

# Add a "Label" column to each dataset
pre_flop_df['Label'] = 'Pre-Flop'
normal_df['Label'] = 'Flop'

# Combine the two datasets into one
combined_df = pd.concat([pre_flop_df, normal_df], ignore_index=True)

# Create the violin plot
fig = px.violin(combined_df, x='Label', y='percentage', color='Label',
                template='plotly_dark', box=True,
                title='Percentage Distribution by Label',width=750)

# Show the plot
fig.show()

```

```python
pre_flop_df = pd.read_csv('pre_flop.csv')
normal_df = pd.read_csv('normal1.csv')

# Add a "Label" column to each dataset
pre_flop_df['Label'] = 'Pre-Flop'
normal_df['Label'] = 'Flop'

# Add percentage2 column to pre-flop dataset
pre_flop_df['percentage2'] = pre_flop_df.apply(lambda row: row['percentage']+0.11 if row['percentage']>0.6556 else row['percentage'], axis=1)

# Add a "percentage2" column to normal_df
normal_df['percentage2'] = normal_df['percentage']

# Combine the two datasets into one
combined_df = pd.concat([pre_flop_df, normal_df], ignore_index=True)

# Create the violin plot
fig = px.violin(combined_df, x='Label', y='percentage2', color='Label',
                template='plotly_dark', box=True,
                title='Percentage2 Distribution by Label',width=750)

# Show the plot
fig.show()
```
